<h1>To-Do List

<h3>Functions

```python
#Class to create task objects
class Task:
    def __init__(self, name, priority):
        self.name = name
        self.priority = priority
        self.completed = False

    # This function marks the task as completed
    def mark_completed(self):
        self.completed = True


#list to add tasks
tasks = []


#Function to add tasks
def add_task():
    #inputs
    name = input("Enter task name: ")
    priority = input("Enter priority (High/Medium/Low): ")

    #Creates task object
    new_task = Task(name, priority)

    #add task object to list
    tasks.append(new_task)

    #return message
    print("Task added successfully!\n")


#Fuction to view tasks
def view_task():
    #check if task is available or not
    if len(tasks) == 0:
        print("No tasks available!")
        return

    #if task is pending then prints it
    print("\nPending Tasks:")
    for index, t in enumerate(tasks):
        if not t.completed:
            print(f"{index + 1}. {t.name} - {t.priority}")

    #prints completed tasks
    print("\nCompleted Tasks:")
    for index, t in enumerate(tasks):
        if t.completed:
            print(f"{index + 1}. {t.name} - {t.priority}")


#Function to mark tasks completed
def mark_task_completed():
    view_task()

    if len(tasks) == 0:
        return

    #Get task number and index
    task_number = int(input("Enter task number to mark as completed: "))
    index = task_number - 1

    #check if index is valid and mark it as complete
    if 0 <= index < len(tasks):
        tasks[index].mark_completed()
        print("Task completed successfully!\n")
    else:
        print("Task not available!")


#Delete tasks
def delete_task():
    view_task()

    if len(tasks) == 0:
        return

    #Get task number and index
    task_number = int(input("Enter task number to delete: "))
    index = task_number - 1

    #check for valid index and deletes task
    if 0 <= index < len(tasks):
        tasks.pop(index)
        print("Task deleted successfully!\n")
    else:
        print("Task not available!")
```

<h3> Main menu

```python
while True:
    print("TO-DO LIST MENU")
    print("1. Add Task")
    print("2. View Tasks")
    print("3. Mark Task as Completed")
    print("4. Delete Task")
    print("5. Exit")

    choice = input("Choose an option (1-5): ")

    match choice:
        case "1":
            add_task()
        case "2":
            view_task()
        case "3":
            mark_task_completed()
        case "4":
            delete_task()
        case "5":
            print("Exiting program. Goodbye!")
            break
        case _:
            print("Invalid choice. Try again.\n")
```
