**Calculator Program**

```

#Addition
def add(a,b):
    add=a+b
    return add

#subtraction
def sub(a,b):
    sub=a-b
    return sub

#Division
def div(a,b):
    div=a/b
    return div

#Multiplication
def mul(a,b):
    mul=a*b
    return mul
```

```
#input

a=int(input("Enter the value of first number"))
b=int(input("Enter the value of second number"))

oper=input("Enter the desired operation(+, -, *, /):"+"")
```

```
#Calculator

if oper=="+":
    print(f"a+b={add(a,b)}")
elif oper=="-":
    print(f"a-b={sub(a,b)}")
elif oper=="/":
    print(f"a/b={div(a,b)}")
elif oper=="*":
    print(f"a*b={mul(a,b)}")
else:
    print("Error Try again later")
```
