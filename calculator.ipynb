<h1>Calculator Program</h1>

**Use of Functions** <br>
(Function Defination)

```
#Addition
def add(a,b):
    add=a+b
    return add

#subtraction
def sub(a,b):
    sub=a-b
    return sub

#Division
def div(a,b):
    div=a/b
    return div

#Multiplication
def mul(a,b):
    mul=a*b
    return mul
```

**Use of dictionary**

```
#Operators Dictionary
operators={
    "+":add,
    "-":sub,
    "*":mul,
    "/":div
}
```

**Taking user Input**

```
#input

a=int(input("Enter the value of first number"))
b=int(input("Enter the value of second number"))

oper=input("Enter the desired operation(+, -, *, /):"+"")
```

<br>(if else statement)

```
#Calculator
if oper in operators:
   print(f"a{oper}b={operators[operator](a,b)}")
else:
   print("Error Try again later")
```
